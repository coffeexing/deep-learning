```python
import numpy as np
import matplotlib.pylab as plt  
```

Step function with NumPy

```python
def step_function(x):
    x = np.array(x)
    y = x > 0
    return y.astype(int)

step_function(3.0), step_function(np.array([1.0, 2.0]))
```

The image of step function

```python
x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

Sigmoid function

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.array([-1.0, 1.0, 2.0])
sigmoid(x)
```

The image of sigmoid function

```python
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

ReLU function

```python
def relu(x):
    return np.maximum(0, x)
```

The image of ReLU function

```python
x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)
plt = plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```
